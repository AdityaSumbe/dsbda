```python
import requests
import bs4
```

```python
url = "https://www.flipkart.com/google-pixel-8a-porcelain-128-gb/p/itm619c078bd72f7"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}
```

```python
request1 = requests.get(url, headers=headers)
print(request1)
```

```python
soup=bs4.BeautifulSoup(request1.content,'html.parser')
print(soup.title.text)
```

```python
# REVIEWS
reviews = soup.find_all('div')
for review in reviews:
    if review.get_text().strip():
        print(review.get_text())
```

```python
#SINGLE REVIEW
rreviews = soup.find('div').get_text()
print(rreviews)
```

```python
#FETCHING RATTINGS
ratings = soup.find_all('div')

for rating in ratings:
    if "★" in rating.get_text():
        print(rating.get_text())
```

```python
#FETCHING CUSTOMER NAMES
print("Customer Names:")
names = soup.find_all('p', class_='review-author')
for name in names:
    print("-", name.get_text().strip())
```

```python
#FETCHING PRODUCT NAME
print("Product name:")
product_name = soup.find('title')
print(product_name.get_text().split('|')[0])
```

```python
#FETCHING PRICE OF PRODUCT
print("Product price:")
for price in soup.find_all(string=True):
    if "₹" in price:
        print(price)
        break
```

```python
#FETCHING QUESTIONS
print("Customer Questions:")
for q in soup.find_all(string=True):
    q = q.strip()
    if q.endswith('?') and len(q) < 120:
        print("-", q)
```
